```python
import re

# Dictionary of patterns and their replacements
patterns_to_replace = {
    r'M[\s.]?Chinnaswamy Stadium.*': 'M. Chinnaswamy Stadium',
    r'Wankhede.*Stadium.*': 'Wankhede International Stadium',
    r'Feroz Shah Kotla.*': 'Arun Jaitley Stadium',
    r'Arun Jaitley Stadium.*': 'Arun Jaitley Stadium',
    r'Eden.*Garden.*': 'Enden Garden',
    r'Dr DY Patil Sports Academy.*':'Dr DY Patil Sports Academy',
    r'Rajiv Gandhi International Stadium.*': 'Rajiv Gandhi International Stadium',
    r'MA Chidambaram Stadium.*': 'MA Chidambaram Stadium'
    
}

# Iterate through the dictionary and apply the replacements
for pattern, replacement in patterns_to_replace.items():
    data['venue'] = data['venue'].replace(to_replace=pattern, value=replacement, regex=True)
```
