## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare some data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (5,5)) # (width, hight)

# 3. plot the data
ax.plot(x, y) # adding data to our axes

# 4. Customize plot
ax.set(title = "Simple-plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Two options for subplots

```python
# subplots option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(20), np.random.random(20));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(5,6));
```

```python
# subplots option 2:
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))

ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(20), np.random.random(20));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.randn(5,6));
```

### Which should you use? (pyplot vs matplotlib OO method?)
* When plotting something quickly,okay to use the pyplot method
* When plotting something more advancd,use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# pyplot method
over_50.plot(x = "age",
             y = "chol",
             c = "target",
             kind = "scatter");
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(x = "age",
             y = "chol",
             c = "target",
             kind = "scatter",
             ax = ax);
# ax.set_xlim([45,100]);
```

```python
over_50.target.values
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

# Customize the plot
ax.set(title = "Heart disease and Cholesterol level",
       xlabel = "Age",
       ylabel = "Cholestarol")

# Add a Legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line
ax.axhline(y = over_50["chol"].mean(), linestyle = '--', linewidth = 2, color = 'r');
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10,10),
                               sharex = True)
# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c= over_50["target"])

# Customize the ax0
ax0.set(title = "Heart Disease & Cholesterol levels",
        ylabel = "Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a mean line
ax0.axhline(y = over_50["chol"].mean(), linestyle = '--', linewidth = 2 , color = 'r');

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])
# Customize the ax1
ax1.set(title = "Heart Disease & Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")

# Add mean line
ax1.axhline(over_50["thalach"].mean(), linestyle = '--', linewidth = 2, color = 'r');

# Add a title to the figure
fig.suptitle("Heart-Disease-Analysis", fontsize = 16, fontweight = "bold", color = 'r');
```

```python
# Set the style
plt.style.use('seaborn-v0_8-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = "winter")

# Customize the plot
ax.set(title = "Heart disease and Cholesterol level",
       xlabel = "Age",
       ylabel = "Cholestarol")

# Change x axis limitations
ax.set_xlim([50, 80])

# Add a Legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line
ax.axhline(y = over_50["chol"].mean(), linestyle = '--', linewidth = 2, color = 'r');
```

```python
# Customizing x and y axis limitation

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10,10),
                               sharex = True)
# Add data to ax0
scatter = ax0.scatter(x = over_50["age"],
                      y = over_50["chol"],
                      c= over_50["target"],
                      cmap = 'winter')

# Customize the ax0
ax0.set(title = "Heart Disease & Cholesterol levels",
        ylabel = "Cholesterol")

# Change the ax0's x axis limitation
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a mean line
ax0.axhline(y = over_50["chol"].mean(), linestyle = '--', linewidth = 2 , color = 'r');

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = 'winter')
# Customize the ax1
ax1.set(title = "Heart Disease & Max Heart Rate",
        xlabel = "Age",
        ylabel = "Max Heart Rate")

# Change ax1's y axis limitation
ax1.set_ylim([60,200])


# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target")

# Add mean line
ax1.axhline(over_50["thalach"].mean(), linestyle = '--', linewidth = 2, color = 'r')

# Add a title to the figure
fig.suptitle("Heart-Disease-Analysis", fontsize = 16, fontweight = "bold", color = 'r')

# Save the figure
fig.savefig("images/Heart-Disease-Analysis.png");
```
